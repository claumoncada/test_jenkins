```python
import pandas as pd
import random
import time
import requests

# Leer el archivo CSV con pandas
ratings_df = pd.read_csv('ratings.csv')

# Obtener la lista de IDs válidos (userId)
valid_ids = ratings_df['userId'].unique()
```

```python
def enviar_peticion(user_id):
    try:
        # Aquí reemplaza con la URL de tu servidor Flask local
        url = "http://127.0.0.1:5000/recommend"

        # Enviar la petición a la API Flask (suponiendo que es una POST request)
        response = requests.post(url, data={"user_id": user_id, "model": "knn"})  # Cambia "knn" por "reg" si quieres probar con el otro modelo

        # Verificar la respuesta del servidor
        if response.status_code == 200:
            print(f"Petición exitosa para el ID {user_id}")
            return True
        else:
            print(f"Error en la petición para el ID {user_id}: {response.status_code}")
            return False
    except requests.exceptions.RequestException as e:
        print(f"Error de conexión o timeout para el ID {user_id}: {e}")
        return False
```

```python
# Función que simula el proceso de envío de IDs válidos e inválidos
def simular_envios(valid_ids):
    fallos = []
    
    # Dividir los IDs válidos en bloques de 25
    bloques = [valid_ids[i:i + 25] for i in range(0, len(valid_ids), 25)]
    
    for bloque in bloques:
        # Enviar los primeros 25 IDs válidos
        print(f"Enviando IDs válidos: {bloque}")
        for user_id in bloque:
            exito = enviar_peticion(user_id)
            if not exito:
                fallos.append(user_id)
            time.sleep(0.5)  # Simular un pequeño retraso entre las peticiones

        # Generar un ID inválido (fuera de los IDs válidos)
        id_invalido = random.randint(min(valid_ids)-100, max(valid_ids)+1000)
        print(f"Enviando un ID inválido: {id_invalido}")
        exito = enviar_peticion(id_invalido)
        if not exito:
            fallos.append(id_invalido)
        time.sleep(0.5)  # Simular un pequeño retraso entre las peticiones

    print("\nSimulación terminada.")
    print(f"Total de fallos: {len(fallos)}")
    print(f"IDs con fallos: {fallos}")

# Llamar a la función de simulación
simular_envios(valid_ids)
```
